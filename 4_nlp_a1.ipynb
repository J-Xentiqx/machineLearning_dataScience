```python
from transformers import pipeline
import pymupdf

doc = pymupdf.open("data/paper.pdf") # open a document
text = ""
for page in doc: # iterate the document pages
  text += page.get_text() # get plain text encoded as UTF-8


summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device="mps")

# 4) Zusammenfassen (kurz & simpel):
result = summarizer(text, max_length=500, min_length=80, do_sample=False)
print(result[0]["summary_text"])



```
